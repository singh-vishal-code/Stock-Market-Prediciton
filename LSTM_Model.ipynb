```python
import yfinance as yf
import pandas as pd
import time

def download_multiple_tickers(tickers):
    start_date = "2015-03-01"
    end_date = "2025-07-01"

    for ticker in tickers:
        file_name = f"{ticker}.csv"
        print(f"\nDownloading data for {ticker}...")

        attempt = 0
        max_retries = 3

        while attempt < max_retries:
            try:
                data = yf.download(ticker, start=start_date, end=end_date)

                if data.empty:
                    print(f" No data found for {ticker}. Skipping.")
                    break

                # Reset index
                data.reset_index(inplace=True)

                # Save to CSV
                data.to_csv(file_name, index=False)
                print(f" Saved to {file_name}")
                break

            except Exception as e:
                print(f"Error downloading {ticker}: {e}")
                if "Rate limited" in str(e) or "Too Many Requests" in str(e):
                    attempt += 1
                    wait_time = 10 * attempt
                    print(f" Rate limit hit. Retrying after {wait_time} seconds...")
                    time.sleep(wait_time)
                else:
                    break
        else:
            print(f" Failed to download {ticker} after {max_retries} attempts.")
```

```python
if __name__ == "__main__":
    tickers = ["NKE", "GS", "JNJ", "PFE", "JPM"]
    download_multiple_tickers(tickers)
```

# NIKE

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator

import matplotlib.pyplot as plt

# ------------------------------
# Step 1: Load & Clean Data
# ------------------------------
df = pd.read_csv('/content/NKE.csv', parse_dates=['Date'])
df.sort_values('Date', inplace=True)

# Convert columns to numeric
numeric_cols = ['Open', 'High', 'Low', 'Close', 'Volume']
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col].astype(str).str.replace(',', '').str.replace('$', ''), errors='coerce')

df.dropna(subset=numeric_cols, inplace=True)
df.reset_index(drop=True, inplace=True)

# ------------------------------
# Step 2: Feature Engineering
# ------------------------------
df['H-L']     = df['High'] - df['Low']
df['O-C']     = df['Open'] - df['Close']
df['7D MA']   = df['Close'].rolling(window=7).mean()
df['14D MA']  = df['Close'].rolling(window=14).mean()
df['21D MA']  = df['Close'].rolling(window=21).mean()
df['7D STD']  = df['Close'].rolling(window=7).std()
df.dropna(inplace=True)

# Select final feature columns
features = ['Close','H-L', 'O-C', '7D MA', '14D MA', '21D MA', '7D STD', 'Volume']
target_col = 'Close'

# ------------------------------
# Step 3: Train-Test Split
# ------------------------------
train_data = df[df['Date'] < '2024-01-01']
test_data  = df[df['Date'] >= '2024-01-01']

scaler = MinMaxScaler()

X_train = scaler.fit_transform(train_data[features])
X_test  = scaler.transform(test_data[features])

y_train = X_train[:, 0]  # scaled 'Close' as target
y_test  = X_test[:, 0]

# Reshape target
y_train = y_train.reshape(-1, 1)
y_test = y_test.reshape(-1, 1)

# ------------------------------
# Step 4: Create TimeSeries Generators
# ------------------------------
look_back = 90
n_features = len(features)

train_generator = TimeseriesGenerator(X_train, y_train, length=look_back, batch_size=10)
test_generator  = TimeseriesGenerator(X_test, y_test, length=look_back, batch_size=10)

# ------------------------------
# Step 5: LSTM Model
# ------------------------------
model = Sequential()
model.add(LSTM(64, activation='relu', input_shape=(look_back, n_features)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
model.fit(train_generator, epochs=50, verbose=0)

# ------------------------------
# Step 6: Predictions
# ------------------------------
test_predictions = model.predict(test_generator)

# Inverse scaling for final 'Close' predictions
# Use only 'Close' column scaler (1st column)
close_scaler = MinMaxScaler()
close_scaler.fit(train_data[['Close']])  # fit on unscaled Close for inverse transform

predicted_close = close_scaler.inverse_transform(test_predictions)
actual_close    = close_scaler.inverse_transform(y_test[look_back:])

# ------------------------------
# Step 7: Evaluation
# ------------------------------
def mean_bias_error(y_true, y_pred):
    return np.mean(y_pred - y_true)

rmse = np.sqrt(mean_squared_error(actual_close, predicted_close))
mape = mean_absolute_percentage_error(actual_close, predicted_close) * 100
mbe  = mean_bias_error(actual_close, predicted_close)

print(f" RMSE: {rmse:.2f}")
print(f" MAPE: {mape:.2f}%")
print(f" MBE : {mbe:.2f}")

# ------------------------------
# Step 8: Plot Results
# ------------------------------
plot_dates = test_data['Date'].iloc[look_back:]

plt.figure(figsize=(16, 6))
plt.plot(plot_dates, actual_close, label='Actual Close Price', c='orange')
plt.plot(plot_dates, predicted_close, label='Predicted Close Price', c='green')
plt.title('Nike (NKE)')
plt.xlabel('Date')
plt.ylabel('Close Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator

import matplotlib.pyplot as plt

# ------------------------------
# Step 1: Load & Clean Data
# ------------------------------
df = pd.read_csv('/content/JPM.csv', parse_dates=['Date'])
df.sort_values('Date', inplace=True)

# Convert columns to numeric
numeric_cols = ['Open', 'High', 'Low', 'Close', 'Volume']
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col].astype(str).str.replace(',', '').str.replace('$', ''), errors='coerce')

df.dropna(subset=numeric_cols, inplace=True)
df.reset_index(drop=True, inplace=True)

# ------------------------------
# Step 2: Feature Engineering
# ------------------------------
df['H-L']     = df['High'] - df['Low']
df['O-C']     = df['Open'] - df['Close']
df['7D MA']   = df['Close'].rolling(window=7).mean()
df['14D MA']  = df['Close'].rolling(window=14).mean()
df['21D MA']  = df['Close'].rolling(window=21).mean()
df['7D STD']  = df['Close'].rolling(window=7).std()
df.dropna(inplace=True)

# Select final feature columns
features = ['Close','H-L', 'O-C', '7D MA', '14D MA', '21D MA', '7D STD', 'Volume']
target_col = 'Close'

# ------------------------------
# Step 3: Train-Test Split
# ------------------------------
train_data = df[df['Date'] < '2024-01-01']
test_data  = df[df['Date'] >= '2024-01-01']

scaler = MinMaxScaler()

X_train = scaler.fit_transform(train_data[features])
X_test  = scaler.transform(test_data[features])

y_train = X_train[:, 0]  # scaled 'Close' as target
y_test  = X_test[:, 0]

# Reshape target
y_train = y_train.reshape(-1, 1)
y_test = y_test.reshape(-1, 1)

# ------------------------------
# Step 4: Create TimeSeries Generators
# ------------------------------
look_back = 90
n_features = len(features)

train_generator = TimeseriesGenerator(X_train, y_train, length=look_back, batch_size=10)
test_generator  = TimeseriesGenerator(X_test, y_test, length=look_back, batch_size=10)

# ------------------------------
# Step 5: LSTM Model
# ------------------------------
model = Sequential()
model.add(LSTM(64, activation='relu', input_shape=(look_back, n_features)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
model.fit(train_generator, epochs=50, verbose=0)

# ------------------------------
# Step 6: Predictions
# ------------------------------
test_predictions = model.predict(test_generator)

# Inverse scaling for final 'Close' predictions
# Use only 'Close' column scaler (1st column)
close_scaler = MinMaxScaler()
close_scaler.fit(train_data[['Close']])  # fit on unscaled Close for inverse transform

predicted_close = close_scaler.inverse_transform(test_predictions)
actual_close    = close_scaler.inverse_transform(y_test[look_back:])

# ------------------------------
# Step 7: Evaluation
# ------------------------------
def mean_bias_error(y_true, y_pred):
    return np.mean(y_pred - y_true)

rmse = np.sqrt(mean_squared_error(actual_close, predicted_close))
mape = mean_absolute_percentage_error(actual_close, predicted_close) * 100
mbe  = mean_bias_error(actual_close, predicted_close)

print(f" RMSE: {rmse:.2f}")
print(f" MAPE: {mape:.2f}%")
print(f" MBE : {mbe:.2f}")

# ------------------------------
# Step 8: Plot Results
# ------------------------------
plot_dates = test_data['Date'].iloc[look_back:]

plt.figure(figsize=(16, 6))
plt.plot(plot_dates, actual_close, label='Actual Close Price', c='orange')
plt.plot(plot_dates, predicted_close, label='Predicted Close Price', c='green')
plt.title('JP Morgan (JPM)')
plt.xlabel('Date')
plt.ylabel('Close Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```
